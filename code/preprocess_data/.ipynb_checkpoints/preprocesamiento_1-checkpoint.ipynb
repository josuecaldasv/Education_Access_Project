```python
path = r'..\..\output\data_build\*.dta'
```

```python
bases = fun.importar_bases( path )
for i, base in enumerate( bases ):
    print( i )
```

## 1.3. ÚLtima limpieza

```python
for i, base in enumerate( bases ):
    data_cols = [col for col in bases[ i ].columns if col not in vb.non_pred_vars ]
    bases[ i ] = bases[ i ][ data_cols ]
    bases[ i ] = bases[ i ].astype( str ).replace( '', np.nan, regex = True )
    bases[ i ] = bases[ i ].astype( str ).replace( '\.+$', np.nan, regex = True ).astype( float )   
```

## 1.4. Duplicar las bases y asignar nombres

```python
dfs = [ pd.DataFrame() for x in range( 2 ) ]
for base in bases:
    dfs = bases.copy()
```

```python
dfs[0].name = 'base0'
dfs[1].name = 'base1'
```
